#Data Visualisasi

```python
# Menghitung total pengguna sepeda per tahun
yearly_day_data = day_df.groupby('yr')['cnt'].sum().reset_index()

# Visualisasi
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_day_data, x='yr', y='cnt', marker='o')
plt.title('Total Penggunaan Sepeda per Tahun (DAY)')
plt.xlabel('Tahun')
plt.ylabel('Total Pengguna')
plt.xticks([0, 1], ['2011', '2012'])
plt.show()
```

```python
# Menghitung total pengguna sepeda per tahun
yearly_hour_data = hour_df.groupby('yr')['cnt'].sum().reset_index()

# Visualisasi
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_hour_data, x='yr', y='cnt', marker='o')
plt.title('Total Penggunaan Sepeda per Tahun (HOUR)')
plt.xlabel('Tahun')
plt.ylabel('Total Pengguna')
plt.xticks([0, 1], ['2011', '2012'])
plt.show()
```

**Baik harian atau jam tiap tahunnya pengguna memiliki kenaikan yang cukup tinggi, sehingga dapat dilihat banyaknya pengguna sepeda per tahunnya**

```python
# Menghitung rata-rata harian pengguna sepeda per tahun
yearly_day_avg = day_df.groupby('yr')['cnt'].mean().reset_index()

# Visualisasi
plt.figure(figsize=(10, 6))
sns.barplot(data=yearly_day_avg, x='yr', y='cnt')
plt.title('Rata-rata Harian Penggunaan Sepeda per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Pengguna Harian')
plt.xticks([0, 1], ['2011', '2012'])
plt.show()
```

**Rata-rata pengguna sepeda terhitung lebih banyak di 2012 sebanyak 5000 lebih pengguna**

```python
# Menghitung total pengguna sepeda per musim
season_data = day_df.groupby('season')['cnt'].sum().reset_index()

# Mapping untuk nama musim
season_map = {1:'springer', 2:'summer', 3:'fall', 4:'winter'}
season_data['season'] = season_data['season'].map(season_map)

# Visualisasi
plt.figure(figsize=(10, 6))
sns.barplot(data=season_data, x='season', y='cnt')
plt.title('Total Penggunaan Sepeda per Musim')
plt.xlabel('Musim')
plt.ylabel('Total Pengguna')
plt.show()
```

```python
# Menghitung total pengguna casual dan registered per musim
season_users = day_df.groupby('season')[['casual', 'registered']].sum().reset_index()
season_users['season'] = season_users['season'].map(season_map)

# Visualisasi
plt.figure(figsize=(10, 6))
sns.barplot(data=season_users.melt(id_vars='season'), x='season', y='value', hue='variable')
plt.title('Perbandingan Pengguna Casual vs Registered per Musim')
plt.xlabel('Musim')
plt.ylabel('Total Pengguna')
plt.legend(title='Tipe Pengguna', loc='upper left')
plt.show()
```

**Jika dilihat pada data grafik total pengguna baik casual atau registered, banyak yang menggunakan sepeda saat musim fall karena berdasarkan kebiasaan orang yang tinggal di negara 4 musim ini musim tersebut sangat cocok untuk melakukan kegiatan bersepeda ketimbang saat musim springer atau winter**
